```python
import pandas as pd
import numpy as np
import math
from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt

import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
x = df['x']
y = df['y']
```

<h1 style= "color: green">Gradient Decent</h1>

<h3 style= "color: indigo">Exercise 1 (learning_rate: .0001, iterations: 500)</h3>

```python
m1 = c1 = 0
l = .0001 #learning_rate
n = len(x)

for i in range(500):
    ypred =( m1*x) + c1
    Dm = (-2/n)* sum(x*(y-ypred))
    Dc = (-2/n)* sum((y-ypred))
    
    #update
    m1 = m1 - (l*Dm)
    c1 = c1 - (l*Dc)
    print(m1,c1)
```

```python
m1,c1
```

```python
48.955889 * m1 + c1
```

<h3 style= "color: indigo">Exercise 2 (learning_rate: .0001, iterations: 1000)</h3>

```python
m2 = c2 = 0
l = .0001 #learning_rate
n = len(x)
cost_previous = 0
for i in range(1000):
    ypred =( m2*x) + c2
    cost = (1/n)*sum((y-ypred)**2)
    Dm = (-2/n)* sum(x*(y-ypred))
    Dc = (-2/n)* sum((y-ypred))
    
    #update
    m2 = m2 - (l*Dm)
    c2 = c2 - (l*Dc)
    print(i,' ',m2,' ',c2, ' ',cost)
```

```python
m2,c2
```

```python
48.955889 * m2 + c2
```

<h3 style= "color: indigo">Exercise 3 (learning_rate: .0003, iterations: untill cost difference .000001)</h3>

```python
m3 = c3 = 0
l = .0003
n = len(x)
cost_previous = 0
for i in range(10000):
    ypred =( m3*x) + c3
    cost = (1/n)*sum((y-ypred)**2)
    Dm = (-2/n)* sum(x*(y-ypred))
    Dc = (-2/n)* sum((y-ypred))
    
    #update
    m3 = m3 - (l*Dm)
    c3 = c3 - (l*Dc)
    d = cost_previous - cost
    if math.isclose(cost, cost_previous, rel_tol=1e-6): #import math
        break
    cost_previous = cost
    print("iteration: {},' ',m: {},' ',c: {},' ',cost: {}, d: {}".format(i, m3, c3, cost, d))
```

<h2 style= "color: green">Visualization of Gradinet Decent</h2>

```python
def gradient_descent(x,y):
    m = c = 0
    rate = 0.0001
    n = len(x)
    plt.scatter(x,y,color='red')
    for i in range(10000):
        y_pred = m * x + c
        plt.plot(x,y_pred,color='green')
        der_m = -(2/n)*sum(x*(y-y_pred))
        der_c = -(2/n)*sum(y-y_pred)
        m = m - rate * der_m
        c = c - rate * der_c
```

```python
x = np.array(df['x'])
y = np.array(df['y'])

gradient_descent(x,y)
```
